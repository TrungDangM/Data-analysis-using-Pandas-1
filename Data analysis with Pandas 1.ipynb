## Question 10: In every Olympics, which US athlete has won the most total number of medals? Include the althelet discipline

```python
First, filter out all of American athlete
```

```python
gy = oo[oo.NOC == 'USA']
gy.head()
```

```python
gy = gy.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value= 0)
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
```

```python
gy.reset_index(inplace = True)
```

```python
for year, group in gy.groupby('Edition'):
    print(group)
```

```python
for year, group in gy.groupby('Edition'):
    print(group.sort_values('Total', ascending = False)[:1]) 
```

```python
tu = [group.sort_values('Total', ascending = False)[:1] for year, group in gy.groupby('Edition')]
tu
```

```python
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
